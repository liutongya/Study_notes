```python
from intake import open_catalog
from matplotlib import pyplot as plt
import numpy as np

cat = open_catalog("https://raw.githubusercontent.com/pangeo-data/pangeo-datastore/master/intake-catalogs/ocean/llc4320.yaml")
ds  = cat["LLC4320_grid"].to_dask()

ds1 = cat["LLC4320_SSH"].to_dask()
ds2 = cat["LLC4320_SST"].to_dask()
```

```python
ds2.SST[-1, 9, :, :].plot.contourf(levels=50, vmin=-6, vmax=2, cmap='magma')
```

```python
lon = ds.XC[9, 1200:1550, 1400:1800].values
lat = ds.YC[9, 1200:1550, 1400:1800].values

nn = 10
for tt in np.arange(-6174,-5429,24):

    fig = plt.figure(figsize=(14, 8))

    plt.subplot(121)
    lev1 = np.linspace(-3, 0, 100)
    data1 = ds1.Eta[tt, 9, 1200:1550, 1400:1800].values
    plt.contourf(lon, lat, data1, lev1, cmap='jet', extend='both')
    c_ticks = [-3, -2, -1, 0]
    plt.colorbar(orientation='horizontal', ticks=c_ticks)
    plt.xlabel('LON')
    plt.ylabel('LAT')
    plt.title('SSH')
    plt.text(172.3, -69.3, ds1.time[tt].values, size=18)

    plt.subplot(122)
    lev2 = np.linspace(-1.95, -1.75, 100)
    data2 = ds2.SST[tt, 9, 1200:1550, 1400:1800].values

    plt.contourf(lon, lat, data2, lev2, cmap='jet', extend='both')
    c_ticks = [-1.9, -1.85, -1.8]
    plt.colorbar(orientation='horizontal', ticks=c_ticks)
    plt.xlabel('LON')
    plt.title('SST')

    name = 'fig_' + str(nn) + '.png'

    plt.savefig(name, dpi=200, bbox_inches='tight')
    plt.close()
    nn = nn + 1
```
